```python
import pandas as pd
import numpy as np
import plotly.offline as plt
import plotly.graph_objs as go

# pd.set_option('display.max_columns', 10)


def k_means_clustering(path, k):
    data = pd.read_csv(path)
#     print('Initial means:\n', k_means)
    var_list = []
    clustersInIterations = pd.DataFrame()
    for x in range(50):
        k_means = (data.sample(k))
        k_means2 = pd.DataFrame()
        clusters = pd.DataFrame()
        while not k_means2.equals(k_means):

            # distance matrix
            cluster_count = 0
            for idx, k_mean in k_means.iterrows():
                clusters[cluster_count] = (data[k_means.columns] - np.array(k_mean)).pow(2).sum(1).pow(0.5)
                cluster_count += 1

            # update cluster
            data['MDCluster'] = clusters.idxmin(axis=1)

            # store previous cluster
            k_means2 = k_means
            k_means = pd.DataFrame()
            k_means = data.groupby('MDCluster').agg(np.mean)

#             print(k_means.equals(k_means2))
            
        var_list.append(np.sum(data.groupby('MDCluster').agg(np.var).pow(2).sum(1).pow(0.5)))
        clustersInIterations[x] = data['MDCluster']
        
    min_var_iter_num = np.argmin(var_list)
    print(clustersInIterations)
    print(var_list)
    data['MDCluster'] = clustersInIterations[min_var_iter_num]
    k_means = pd.DataFrame()
    k_means = data.groupby('MDCluster').agg(np.mean)
    
    
    # plotting
    print('Plotting...')
    data_graph = [go.Scatter(
        x=data['V1'],
        y=data['V2'].where(data['MDCluster'] == c),
        mode='markers',
        name='Cluster: ' + str(c)
    ) for c in range(k)]

    data_graph.append(
        go.Scatter(
            x=k_means['V1'],
            y=k_means['V2'],
            mode='markers',
            marker=dict(
                size=10,
                color='#000000',
            ),
            name='Centroids of Clusters'
        )
    )

    plt.plot(data_graph, filename='cluster.html')

k_means_clustering(path='k_means_clustering_test_1.csv', k=7)
```
